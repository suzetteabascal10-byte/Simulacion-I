# Prueba de Kolmogorov-Smirnov
Se utilizaran una muestra de numeros aleatorios así como se orderara de forma ascendente así como calcularemos el estadistico de prueba D

Despues de ordenar los datos agregamos la coomparativa con el estadistico D y el valor critico d

```python
import pandas as pd
import math

data = {
    'x_i': t,
    '(i-1)/n': v,
    'i/n': b,
    'D^-': r,
    'D^+': s
}

# Crear la tabla
df = pd.DataFrame(data)

# Imprimir la tabla
print(df)

# Nivel de significancia
a = 0.05

# Valor crítico aproximado:
c_a = math.sqrt(-0.5 * math.log(a/2))
d_c = c_a / math.sqrt(n)

if D_ks < d_c:
    print("Como D < d_c, no se rechaza H0 ")
else:
    print("Como D >= d_c, se rechaza H0 ")
```
